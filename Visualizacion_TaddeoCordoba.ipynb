```python
# Resumen estadístico de las edades por género
summary = df.groupby('Gender')['Age'].describe()

print(summary)
```

```python
# Crear un gráfico de barras de la media de edad por género
bars = plt.bar(summary.index, summary['mean'], color=['blue', 'pink'])
plt.xlabel('Género')
plt.ylabel('Media de Edad')
plt.title('Media de Edad por Género')

# Agregar etiquetas a las barras
for bar, mean_age in zip(bars, summary['mean']):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, mean_age + 0.5, f'{mean_age:.2f}', ha='center', color='black', fontsize=10)

plt.show()
```

```python
# Calcular el promedio de pasajeros por continente
average_passengers = df.groupby('Continents')['Passenger ID'].mean()

print (average_passengers)
```

```python
# Promedio de Pasajeros por Continente

#Crear colores diferentes para cada barra
colors = plt.cm.viridis(np.linspace(0, 1, len(average_passengers)))

# Crear un gráfico de barras agrupadas con colores y etiquetas
plt.figure(figsize=(10, 6))
bars = plt.bar(average_passengers.index, average_passengers.values, color=colors)

# Agregar etiquetas en cada barra
for bar, label in zip(bars, average_passengers.values):
    plt.text(bar.get_x() + bar.get_width() / 2, label, f'{label:.2f}', ha='center', va='bottom', fontsize=10)

plt.xlabel('Continente')
plt.ylabel('Promedio de Pasajeros')
plt.title('Promedio de Pasajeros por Continente')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje X para mayor claridad
plt.show()
```

En esta gráfica que muestra el promedio de pasajeros por continente, se puedee realizar varios análisis e interpretaciones:

1. **Comparación de Promedios**: se compara visualmente los promedios de pasajeros en cada continente representado por las barras. 

2. **Distribución Relativa**: Observando la longitud de las barras, se puede hacer una estimación de la distribución relativa de pasajeros por continente. ¿Cuál es el continente que contribuye significativamente a la cantidad total de pasajeros? ¿Hay algún continente con un promedio de pasajeros notablemente diferente de los demás?

3. **Identificación de Outliers**: Se identifican que continentes se destacan como "outliers" si tienen promedios de pasajeros significativamente más altos o más bajos que los demás. Podemos observar que los promedios de pasajeros por continentes son relativamente iguales.

5. **Tendencias Geográficas**: La gráfica puede proporcionar información sobre tendencias geográficas o económicas. Por ejemplo, si un continente tiene un promedio de pasajeros más alto, esto podría relacionarse con su tamaño geográfico, población o desarrollo económico.

6. **Decisiones de Negocio**: Si estámo tomando decisiones relacionadas con viajes o transporte en función de estos datos, la gráfica nos ayuda a identificar áreas geográficas que podrían requerir más atención o inversión.

En general, esta gráfica de barras permite resumir y visualizar de manera efectiva la distribución de pasajeros por continente, lo que puede ser útil para tomar decisiones informadas y comprender mejor los datos.

```python
#Histograma de Estado de Vuelvo a lo largo del tiempo
# Convertir la columna "Departure Date" a tipo datetime
df['Departure Date'] = pd.to_datetime(df['Departure Date'])

# Extraer el mes y el año de la columna "Departure Date"
df['Month'] = df['Departure Date'].dt.month
df['Year'] = df['Departure Date'].dt.year

# Crear una tabla pivote para contar la frecuencia de cada estado de vuelo por mes y año
pivot_table = df.pivot_table(index=['Year', 'Month'], columns='Flight Status', aggfunc='size', fill_value=0)

# Reiniciar el índice de la tabla pivote
pivot_table = pivot_table.reset_index()

# Crear un gráfico de línea usando Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=pivot_table, x='Month', y='On Time', label='On Time')
sns.lineplot(data=pivot_table, x='Month', y='Delayed', label='Delayed')
sns.lineplot(data=pivot_table, x='Month', y='Cancelled', label='Cancelled')

plt.xlabel('Mes')
plt.ylabel('Frecuencia')
plt.title('Estado de Vuelo a lo largo del tiempo (por Mes y Año)')
plt.legend(title='Flight Status')
plt.show()
```

El análisis de la gráfica que muestra la evolución de los estados de vuelo ("On Time," "Delayed," "Cancelled") a lo largo del tiempo (por mes y año) proporciona información valiosa sobre el desempeño y la tendencia de los vuelos en el período de tiempo representado en los datos. A continuación, se detallan algunas de las observaciones y análisis que puedes realizar:

1. **Tendencia Temporal**: Puedes identificar tendencias a lo largo del tiempo para cada estado de vuelo. Por ejemplo, ¿hay una disminución o un aumento en los vuelos "On Time" a medida que pasa el tiempo? ¿Hay un patrón estacional en los vuelos "Delayed" o "Cancelled"?

2. **Comparación de Estados de Vuelo**: La gráfica permite comparar visualmente la frecuencia de los diferentes estados de vuelo. ¿Qué proporción de vuelos está "On Time" en comparación con los vuelos "Delayed" o "Cancelled"? ¿Hay alguna relación entre estos estados de vuelo a lo largo del tiempo?

3. **Identificación de Problemas**: Si observamos picos en los vuelos "Delayed" o "Cancelled" en ciertos meses o años, esto puede indicar problemas operativos o condiciones especiales en esos períodos que afectaron a los vuelos.

4. **Patrones Estacionales**: Se puede identificar patrones estacionales, como aumento de vuelos "Delayed" durante las temporadas de mal tiempo o vacaciones, o disminución de vuelos "Cancelled" en meses específicos.

5. **Seguimiento de Mejoras**: Si se han implementado cambios o mejoras en la operación de vuelos, puedes observar si estos cambios han tenido un impacto en la puntualidad ("On Time") o en la reducción de vuelos "Cancelled."

6. **Identificación de Necesidades de Recursos**: Si hay un aumento constante en los vuelos "Delayed" o "Cancelled," esto podría indicar la necesidad de asignar más recursos para mantener la puntualidad y reducir las cancelaciones.

En resumen, esta gráfica te permite evaluar la eficiencia y la confiabilidad de los vuelos a lo largo del tiempo y tomar decisiones informadas para mejorar la operación y la satisfacción de los pasajeros.
